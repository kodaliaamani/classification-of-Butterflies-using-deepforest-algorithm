```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
import glob

from sklearn.metrics import classification_report
from deepforest import CascadeForestClassifier
```

```python
directory = '../input/butterfly-images40-species/'
train_dir = os.path.join(directory, 'train')
test_dir = os.path.join(directory, 'test')
val_dir = os.path.join(directory, 'valid')
```

```python
trainImages = []
trainLabels = []
testImages = []
testLabels = []

for file in glob.glob(train_dir + '/*/*' + '.jpg'):
    trainImages.append(cv2.resize(cv2.imread(file, 0), (100,100)))
    trainLabels.append(file.split('/')[-2])

for file in glob.glob(val_dir + '/*/*' + '.jpg'):
    testImages.append(cv2.resize(cv2.imread(file, 0), (100,100)))
    testLabels.append(file.split('/')[-2])
```

```python
len(trainLabels), len(trainImages), len(testImages), len(testLabels)
```

```python
trainImages = np.array(trainImages)
testImages = np.array(testImages)
```

```python
trainImages.shape, testImages.shape
```

```python
plt.figure(figsize=(10, 10))
i = 0
for images, labels in zip(testImages, testLabels):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images)
    plt.title(labels)
    plt.axis("off")
    i += 1
    if i == 9:
        break
```

```python
model = CascadeForestClassifier()
```

```python
model.fit(trainImages.reshape(9285, 10000), trainLabels)
```

```python
y_pred = []
errors = 0
count = 0
preds = model.predict(testImages.reshape(375, 10000))
for i, p in enumerate(preds):
    count += 1
    y_pred.append(p)  
    if p != testLabels[i]:
        errors += 1

print("Accuracy",(count-errors)* 100/count)
print(classification_report(np.array(testLabels), np.array(y_pred), target_names=list(set(trainLabels))))
```

```python
file = "../input/butterfly-images40-species/test/BECKERS WHITE/1.jpg"

img = cv2.resize(cv2.imread(file, 0), (100,100)).reshape(1, 10000)
name = model.predict(img)[0]
```

```python
score =  max(model.predict_proba(img)[0])
```
